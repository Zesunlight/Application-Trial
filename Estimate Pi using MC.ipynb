#  Estimating Value of Pi using Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arc = np.linspace(0, np.pi/2, 100)

plt.axes().set_aspect('equal')
plt.plot(np.cos(arc), np.sin(arc), color="red")

number = 2000
sample = np.random.rand(number, 2)
plt.scatter(sample[:, 0], sample[:, 1], marker=".", color="green");
```

```python
r = np.sqrt(sample[:, 0] ** 2 + sample[:, 1] ** 2)
inside = sum(1 for i in r if i <= 1)
print(4 * inside / number)
```
